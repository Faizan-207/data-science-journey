```python
import pandas as pd 
import seaborn as sns
```

```python
tips = sns.load_dataset('tips')
```

```python
titanic = pd.read_csv('../datasets/train.csv')
```

```python
flights = sns.load_dataset('flights')
```

```python
iris = sns.load_dataset('iris')
```

## Scatterplot (N-N)

```python
sns.scatterplot(x='total_bill',y='tip',data =tips,hue ='sex',style ='time',size ='size')
```

## Barplot

```python
sns.barplot(x='Sex',y='Survived',data = titanic,hue ='Embarked')
```

## Boxplot (N-C)

```python
sns.boxplot(x='Age',y='Sex',data=titanic,hue='Survived')
```

## Distplot (N-C)

```python
survived = titanic [titanic['Survived']==1]
not_survived = titanic [titanic['Survived']==0]
sns.displot(x='Age',data=survived,kind ='kde')
sns.displot(x='Age',data=not_survived,kind ='kde')
```

```python
sns.displot(x='Age',data=titanic,hue='Survived',kind ='kde')
```

## Heatmap (C-C)


```python
sns.heatmap(pd.crosstab(titanic['Survived'],titanic['Sex']))
```


## Pairplot

```python
sns.pairplot(iris,hue='species')
```

## Lineplot

```python
new = flights.groupby('year')['passengers'].sum().reset_index()
new
```

```python
sns.lineplot(x='year',y='passengers',data=new)
```
