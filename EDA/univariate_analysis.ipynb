```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../datasets/train.csv')
```

```python
df.head()
```

# 1. Categorical data

### Countplot 

```python
sns.countplot(x ='Pclass',data =df)
```

### PieChart 

```python
df['Pclass'].value_counts().plot(kind='pie',autopct ='%.2f')
```

#  2.Numerical data

### Histrogram

```python
import matplotlib.pyplot as plt
```

```python
plt.hist(x='Age',data=df,bins=5)
```

### Displot

```python
# Probablity density function
sns.displot(x='Age', data=df, kde=True,stat='density',bins=10)
```

```python
df['Age'].skew()
# < 0 negative skewed
# = 0 symmetric
# > 0 positive skewed
```

## Boxplot

```python
sns.boxplot(x='Age',data=df)
```
